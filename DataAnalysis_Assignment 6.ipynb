```python
import pandas as pd
from matplotlib.pyplot import pie,axis,show
%matplotlib inline

url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"

titanic = pd.read_csv(url)

s=titanic.pclass.groupby(titanic.sex).sum()
axis('equal')
print(s)
pie(s,labels=s.index)
show()
```

```python
from  matplotlib import pyplot
import seaborn
plot = seaborn.FacetGrid(data=titanic, hue='sex', aspect=2.61,size=3)
plot.map(pyplot.scatter, 'age', 'fare').add_legend()
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatterplot")


```
